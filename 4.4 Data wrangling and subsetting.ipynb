# 0.3 Data Wrangling

```python
# Dropping eval_set column from orders.csv
df_ords.drop(columns=['eval_set'])
```

```python
# Overwriting the dataframe
df_ords=df_ords.drop(columns=['eval_set'])
```

```python
# Looking for missing values (NaN) in a specific column
df_ords['days_since_prior_order'].value_counts(dropna=False)
```

```python
# Renaming the order_dow column
df_ords.rename(columns={'order_dow':'orders_day_of_week'}, inplace=True)
df_ords.head()
```

```python
# Changing data type of order_id column
df_ords['order_id']=df_ords['order_id'].astype('str')
df_ords['order_id'].dtype
```

```python
# Transposing df_dep data set
df_dep.T
```

```python
# Creating a new dataframe from the transposed version
df_dep_t=df_dep.T
df_dep_t
```

```python
# Creating an index
df_dep_t.reset_index()
```

```python
# Take the first row of df_dep for the header
new_header=df_dep_t.iloc[0]
new_header
```

```python
# Take the data under the header row for a new df
df_dep_t_new=df_dep_t[1:]
df_dep_t_new
```

```python
# Set the header row as the df header
df_dep_t_new.columns=new_header
df_dep_t_new
```

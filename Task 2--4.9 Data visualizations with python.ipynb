# Scatterplot

```python
# Create a scatterplot to see if there is a connection between age and income
scatterplot_age_income = sns.scatterplot(x = 'Age', y = 'income', data = df_ords_prods_all)
```

The scatterplot reveals that the most common income range across all ages is between $0 and $200,000. Subsequently, incomes ranging from $200,000 to $300,000 are most prevalent among individuals aged 40 and above. Incomes between $200,000 and $400,000 are observed across all age groups. Moreover, incomes within the $300,000 to $600,000 bracket are sporadically distributed among the age groups of 40 and above.

```python
# Exporting the scatterplot
scatterplot_age_income.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'scatterplot_age_income.png'))
```
